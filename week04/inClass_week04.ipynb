```python
# 1. Hard code an option
style = "ggplot"

with plt.style.context(style):
    # CMD + ]
    fig, ax = plt.subplots(figsize=(8,3))
    ax.plot(stats['max'], marker='.', label='Max')
    ax.plot(stats['min'], marker='.', label='Min')
    ax.set_xlabel('Year')
    ax.set_ylabel('Square Footage')
    ax.set_yscale('log')
    ax.legend()
    plt.show()
```

```python
# 2. make hardcoded option an input to a function
#style = "ggplot"

def make_plot(style):
    with plt.style.context(style):
        # CMD + ]
        fig, ax = plt.subplots(figsize=(8,3))
        ax.plot(stats['max'], marker='.', label='Max')
        ax.plot(stats['min'], marker='.', label='Min')
        ax.set_xlabel('Year')
        ax.set_ylabel('Square Footage')
        ax.set_yscale('log')
        ax.legend()
        plt.show()
```
