```python
# 1. Hard code an option
style = "ggplot"

with plt.style.context(style):
    # CMD + ]
    fig, ax = plt.subplots(figsize=(8,3))
    ax.plot(stats['max'], marker='.', label='Max')
    ax.plot(stats['min'], marker='.', label='Min')
    ax.set_xlabel('Year')
    ax.set_ylabel('Square Footage')
    ax.set_yscale('log')
    ax.legend()
    plt.show()
```

```python
# 2. make hardcoded option an input to a function
#style = "ggplot"

def make_plot(style):
    with plt.style.context(style):
        # CMD + ]
        fig, ax = plt.subplots(figsize=(8,3))
        ax.plot(stats['max'], marker='.', label='Max')
        ax.plot(stats['min'], marker='.', label='Min')
        ax.set_xlabel('Year')
        ax.set_ylabel('Square Footage')
        ax.set_yscale('log')
        ax.legend()
        plt.show()
```

```python
# 2. Make a function with the hard-coded values as inputs

#colormap_name = 'terrain'
#take_log = True

def make_mich_plot_fancy(colormap_name,take_log = True):
    if take_log:
        norm = matplotlib.colors.SymLogNorm(10)
    else:
        norm = matplotlib.colors.Normalize()

    plt.imshow(michigan, cmap=colormap_name, norm = norm)
    plt.colorbar()
    plt.clim(-352,352)
    plt.show()
```

```python
# Step 2.5 test out the function
make_mich_plot_fancy('terrain', take_log=True)
```

```python
# 3. Use ipywidgets.interact to make this interactive
@ipywidgets.interact(colormap_name = plt.colormaps(), take_log = True)
def make_mich_plot_fancy(colormap_name,take_log = True):
    if take_log:
        norm = matplotlib.colors.SymLogNorm(10)
    else:
        norm = matplotlib.colors.Normalize()

    plt.imshow(michigan, cmap=colormap_name, norm = norm)
    plt.colorbar()
    plt.clim(-352,352)
    plt.show()
```

```python
button = ipywidgets.Button(description='I am a button.')

out = ipywidgets.Output()

def say_click(event):
    with out:
        print("I have clicked!  Hurray!")
        
button.on_click(say_click)
```

```python
button
```

```python
ipywidgets.VBox([button,out])
```
